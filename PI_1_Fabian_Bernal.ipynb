```python
#Normalizar la columna producto_id a 13 digitos
def normalizar_producto_id(id):
    
    if (type(id) != str):
        return id
    
    cantidad_caracter = len(id)
    if (len(id) < 13):
        id = '0' * (13 - cantidad_caracter) + id

    if ('-' in id):
        id = id[id.rfind('-') + 1: cantidad_caracter]

    if ('.' in id):
        id = id[0:id.find(',')]
    
    return id
```

```python
#Aplicar la función para dejar 13 digitos en la columna 'producto_id'
df_ps_20200413_2['producto_id']= df_ps_20200413_2['producto_id'].apply(normalizar_producto_id)
```

```python
#Función para cambiar el formato de fecha
def func_fecha(cadena):
    if type (cadena) == str:
        return cadena
    elif type (cadena) == dt.datetime:
        cadenasuc = cadena.strftime('%d/%m/%Y')
        cadenanueva = cadenasuc.replace('/','-')
        lista = cadenanueva
        for i in range(0,2):
            listanueva = lista.replace('0','')
        return listanueva
```
